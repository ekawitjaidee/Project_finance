```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime
import finplot as fplt

%matplotlib notebook
```

```python
df =pd.read_csv('Ground Truth file/KO(2010,2011).csv')
# df=df.set_index('Date')
day = datetime.datetime(df['Date'])
```

```python
df['buy'] = np.where(df['Signal'] == 'buy',df['Close'],np.nan)
df['sold'] = np.where(df['Signal'] == 'sell',df['Close'],np.nan)
```

```python
plt.figure(figsize=(10,8))
# plt.grid()
plt.title('Buy and Sold Signal')
plt.plot(df['Close'])
plt.plot(df['buy'],color='yellow')
plt.plot(df['sold'],color='red')
plt.ylabel('Price')
plt.xlabel('Date')
```
